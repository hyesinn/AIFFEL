### 모듈 리스트

```python
# 파일 불러오기
import os

import pandas as pd
import numpy as np

# 데이터 시각화
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import seaborn as sns

# 다중 공선성 확인
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 사용 모델
import xgboost as xgb
import lightgbm as lgb

# 경고 무시
import warnings
warnings.filterwarnings("ignore")
warnings.filterwarnings("ignore", category = DeprecationWarning)

# 데이터 분리
from sklearn.model_selection import train_test_split

# 모델 성능 검증
from sklearn.metrics import mean_squared_error 
from sklearn.model_selection import KFold, cross_val_score

# 정규화
from scipy.special import boxcox1p, inv_boxcox1p
from sklearn.preprocessing import RobustScaler

from IPython.display import Image
```

### 데이터 파일 저장

```python
# 전체 경로
data_dir = os.getenv('HOME') + '/aiffel/kaggle_kakr_housing/data'

# 파일 경로
train_data_path = os.path.join(data_dir, 'train.csv')
test_data_path = os.path.join(data_dir, 'test.csv')

# 파일 변수에 저장
train_data = pd.read_csv(train_data_path)
test_data = pd.read_csv(test_data_path)
```

### 데이터 탐색

[데이터 탐색 과정](https://github.com/hyesinn/AIFFEL/blob/master/%5BE-05%5Dhouse_price_prediction/%5BE-05%5DEDA.ipynb)

### 데이터 정규화

```python
# 타겟 데이터 그래프
fig, ax = plt.subplots(figsize = (10, 5))
sns.kdeplot(train_data['price'])
```

```python
# 타겟 데이터 정규화
train_data['price'] = boxcox1p(train_data['price'], 0.01) 
                               
fig, ax = plt.subplots(figsize = (10, 5))
sns.kdeplot(train_data['price'])
```

- 타겟 데이터 정규화 boxcox 알고리즘을 사용
- log와 차이점: log 변환 -> 역변환은 평균을 반환, boxcox 변환 -> 역변환은 중간값을 반환
<br/>
[Forecasting : Principles and Practice, 변환과 조정](https://otexts.com/fppkr/transformations.html)

### 데이터 변환

```python
# date 형식 object에서 int로 변환 (yyyymmddT000000에서 yyyymmdd로)
train_data['date'] = train_data['date'].apply(lambda x : str(x[:6])).astype(int)
test_data['date'] = test_data['date'].apply(lambda x : str(x[:6])).astype(int)

# 무의미한 id 삭제
del train_data['id']
del test_data['id']

# 훈련 데이터에서 타겟(가격) 분리
y = train_data['price']
del train_data['price']
```

### 다중공선성 확인

```python
# 다중공선성 확인
vif = pd.DataFrame()
vif['Features'] = train_data.columns
vif['vif'] = [variance_inflation_factor(train_data.values, i) for i in range(train_data.shape[1])]

vif.sort_values(by='vif',ascending=False)
```

### 표준 정규화

```python
# 로버스트 스케일링
rbst_scaler=RobustScaler()
train_rbst=rbst_scaler.fit_transform(train_data)
test_rbst=rbst_scaler.transform(test_data)
```

- 다중공선성 (독립 변수 간 깊은 상관관계) 문제로 로버스트 스케일링
<br/>
[토닥토닥 파이썬 - 머신러닝, 표준 정규화 (RobustScaler)](https://wikidocs.net/89704)

### 모델 설계

[하이퍼파라미터 튜닝 과정](https://github.com/hyesinn/AIFFEL/blob/master/%5BE-05%5Dhouse_price_prediction/%5BE-05%5Dhyperparam.ipynb)

```python
# 모델 정의
random_state = 1

xgboost = xgb.XGBRegressor(learning_rate=0.02, max_depth=6, n_estimators=5000, random_state=random_state)
lightgbm = lgb.LGBMRegressor(learning_rate=0.05, max_depth=13, n_estimators=700, num_leaves=31, random_state=random_state)

models = [{'model':xgboost, 'name':'XGBoost'},
          {'model':lightgbm, 'name':'LightGBM'}]
```

### 교차 검증

```python
# 교차 검증 점수
def get_cv_score(models):
    kfold = KFold(n_splits = 5, shuffle = True, random_state = random_state).get_n_splits(train_rbst)
    print("{:<20} {:<20}".format('Model','CV score'))
    for model in models:
        print("{:<20} {:<20}".format(model['name'], np.mean(np.sqrt(-cross_val_score(model['model'], train_rbst, y,scoring="neg_mean_squared_error"))), kf=kfold))
```

```python
get_cv_score(models)
```

### 블렌딩

```python
# 여러 모델의 예측 평균 함수
def AverageBlending(models, train, y, test):
    for model in models:
        model['model'].fit(train, y)
        
    predictions = np.column_stack([inv_boxcox1p(model['model'].predict(test), 0.01) for model in models])
    
    return np.mean(predictions, axis = 1)
```

```python
# 시험 데이터 예측 및 제출 파일 생성 함수
def save_submission(prediction, model_name, n):
    data_dir = os.getenv('HOME') + '/aiffel/kaggle_kakr_housing/data'
    submission_path = os.path.join(data_dir, 'sample_submission.csv')
    submission = pd.read_csv(submission_path)
    submission['price'] = prediction
    submission_csv_path = '{}/submission_{}_{}.csv'.format(data_dir, model_name, n)
    submission.to_csv(submission_csv_path, index = False)
```

### 최종 결과

```python
pred = AverageBlending(models, train_rbst, y, test_rbst)
# 최종 파일 생성
save_submission(pred, 'blending', '21')
```

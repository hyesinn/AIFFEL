```python
import os, re
import random, math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings(action='ignore')
```

**데이터 불러오기**

데이터 경로

```python
# 데이터 경로
ROOT_PATH = os.path.join(os.getenv('HOME'), 'aiffel')
TRAIN_PATH = ROOT_PATH + '/chest_xray/data/train/*/*'
VAL_PATH = ROOT_PATH + '/chest_xray/data/val/*/*'
TEST_PATH = ROOT_PATH + '/chest_xray/data/test/*/*'

# 데이터 파일명 리스트
train_filenames = tf.io.gfile.glob(TRAIN_PATH)
val_filenames = tf.io.gfile.glob(VAL_PATH)
test_filenames = tf.io.gfile.glob(TEST_PATH)

# 데이터 개수 확인
print("{:<5} {:<0}".format('train', len(train_filenames)))
print("{:<5} {:<0}".format('val', len(val_filenames)))
print("{:<5} {:<0}".format('test', len(test_filenames)))
```

훈련용 데이터 검증용 데이터 분리

```python
# train val re-split
train_val_filenames = train_filenames + val_filenames

random.seed(15)
random.shuffle(train_val_filenames)

train_size = int(len(train_val_filenames) * 0.8)
train_filenames = train_val_filenames[:train_size]
val_filenames = train_val_filenames[train_size:]

print("{:<5} {:<0}".format('train', len(train_filenames)))
print("{:<5} {:<0}".format('val', len(val_filenames)))
```

데이터 균형 확인 및 불균형 가중치

```python
# 데이터 균형 확인
COUNT_TRAIN_NORMAL = len([filename for filename in train_filenames if "NORMAL" in filename])
COUNT_TRAIN_PNEUMONIA = len([filename for filename in train_filenames if "PNEUMONIA" in filename])
COUNT_TRAIN_TOTAL = COUNT_TRAIN_NORMAL+COUNT_TRAIN_PNEUMONIA

COUNT_VAL_NORMAL = len([filename for filename in val_filenames if "NORMAL" in filename])
COUNT_VAL_PNEUMONIA = len([filename for filename in val_filenames if "PNEUMONIA" in filename])
COUNT_VAL_TOTAL = COUNT_VAL_NORMAL+COUNT_VAL_PNEUMONIA



print("{:<20} {}/{}".format("Normal in train", COUNT_TRAIN_NORMAL, COUNT_TRAIN_TOTAL))
print("{:<20} {}/{}".format("Pneumonia in train", COUNT_TRAIN_PNEUMONIA, COUNT_TRAIN_TOTAL))
print("{:<20} {}/{}".format("Normal in val", COUNT_VAL_NORMAL, COUNT_VAL_TOTAL))
print("{:<20} {}/{}".format("Pneumonia in val", COUNT_VAL_PNEUMONIA, COUNT_VAL_TOTAL))
```

```python
# 데이터 불균형 가중치
COUNT_TRAIN_TOTAL = COUNT_TRAIN_NORMAL+COUNT_TRAIN_PNEUMONIA
weight_for_0 = (1 / COUNT_TRAIN_NORMAL) * (COUNT_TRAIN_TOTAL / 2)
weight_for_1 = (1 / COUNT_TRAIN_PNEUMONIA) * (COUNT_TRAIN_TOTAL / 2)

class_weight = {0: weight_for_0, 1: weight_for_1}

print('Weight for NORMAL: {:.2f}'.format(weight_for_0))
print('Weight for PNEUMONIA: {:.2f}'.format(weight_for_1))
```

데이터 Filename 데이터셋 활용

```python
# 데이터 파일명 리스트 데이터셋 생성
train_filenames_ds = tf.data.Dataset.from_tensor_slices(train_filenames)
val_filenames_ds = tf.data.Dataset.from_tensor_slices(val_filenames)
test_filenames_ds = tf.data.Dataset.list_files(TEST_PATH)

# 데이터 확인
for filename in train_filenames_ds.take(1):
    print(filename.numpy())
```

데이터 라벨링

```python
# 데이터 라벨링
def get_label(file_path):
    parts = tf.strings.split(file_path, os.path.sep) # 디렉토리명 활용 (끝에서 두번째)
    return parts[-2] == "PNEUMONIA" # 폐렴이면 양성(True), 노말이면 음성(False)
```

**데이터 전처리**

데이터 형식 설정

```python
# 데이터 형식 설정
def decode_img(img, img_size = [180, 180]):
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32) # 형식 float32
    img = tf.image.resize(img, img_size) # 이미지 크기 조정
    return img
```

데이터 로드

```python
# 데이터 불러오기
def process_path(file_path):
    label = get_label(file_path) # 라벨 검출
    img = tf.io.read_file(file_path) # 이미지 읽기
    img = decode_img(img) # 이미지를 알맞은 형식으로 수정
    return img, label


AUTOTUNE = tf.data.experimental.AUTOTUNE # 값을 동적으로 자동 설정
train_ds = train_filenames_ds.map(process_path, num_parallel_calls = AUTOTUNE)
val_ds = val_filenames_ds.map(process_path, num_parallel_calls = AUTOTUNE)
test_ds = test_filenames_ds.map(process_path, num_parallel_calls = AUTOTUNE)

# 데이터 확인
for image, label in train_ds.take(1):
    print("{:>13} {}".format("Image shape:", image.numpy().shape))
    print("{:>13} {}".format("Label:", label.numpy()))
```

Augmentation

```python
# data augmentation

# filp
def flip(image,label):
    image = tf.image.random_flip_up_down(image)  # 랜덤하게 위 아래를 반전합니다.
    return image, label

# contrast
def contrast(image, label):
    image = tf.image.random_contrast(image, 0.7, 1.3) # 랜덤하게 대비를 조정합니다. 
    return image, label

# 적용한 데이터셋을 추가합니다.
train_ds_aug1 = train_ds.map(flip, num_parallel_calls = 2) # flip
train_ds_aug2 = train_ds.map(contrast, num_parallel_calls = 2) # contrast
train_ds_aug3 = train_ds_aug1.map(contrast, num_parallel_calls = 2) # both
```

데이터 로드

```python
# 데이터 준비
batch_size = 16

def prepare_for_training(ds):
    ds = ds.shuffle(buffer_size = 1000)
    ds = ds.repeat()
    ds = ds.batch(batch_size)
    ds = ds.prefetch(buffer_size = AUTOTUNE)
    return ds

train_ds = prepare_for_training(train_ds)
train_ds_aug1 = prepare_for_training(train_ds_aug1)
train_ds_aug2 = prepare_for_training(train_ds_aug2)
train_ds_aug3 = prepare_for_training(train_ds_aug3)
val_ds = prepare_for_training(val_ds)
```

데이터 미리보기

```python
# 데이터 확인 (미리보기)
def show_batch(image_batch, label_batch):
    plt.figure(figsize = (20, 8))
    for n in range(batch_size):
        ax = plt.subplot(4, (batch_size // 2), (n + 1))
        plt.imshow(image_batch[n])
        if label_batch[n]:
            plt.title("P")
        else:
            plt.title("N")
        plt.axis("off")

image_batch, label_batch = next(iter(train_ds))
show_batch(image_batch.numpy(), label_batch.numpy())
```

```python
# 변형 데이터 확인 (미리보기)
image_batch, label_batch = next(iter(train_ds_aug3))
show_batch(image_batch.numpy(), label_batch.numpy())
```

**모델링**

모델링 레이어 블록

```python
# 모델링 레이어 블록

def input_block(img_size = [180, 180]):
    block = tf.keras.Sequential([
        tf.keras.Input(shape = (img_size[0], img_size[1], 3)),
        tf.keras.layers.Conv2D(16, 3, activation = 'relu', padding = 'same'),
        tf.keras.layers.Conv2D(16, 3, activation = 'relu', padding = 'same'),
        tf.keras.layers.MaxPool2D()
    ])
    return block

def conv_block(filters):
    block = tf.keras.Sequential([
        tf.keras.layers.Conv2D(filters, 3, activation = 'relu', padding = 'same'),
        tf.keras.layers.Conv2D(filters, 3, activation = 'relu', padding = 'same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ])
    return block

def dense_block(units, dropout_rate):
    block = tf.keras.Sequential([
        tf.keras.layers.Dense(units, activation = 'relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(dropout_rate)
    ])
    return block

def dense_block2(units, dropout_rate):
    block = tf.keras.Sequential([
        tf.keras.layers.Dense(units, activation = 'relu'),
        # tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(dropout_rate)
    ])
    return block
```

```python
# 모델 1 
def build_model():
    model = tf.keras.Sequential([
        input_block(),
        
        conv_block(32),
        conv_block(64),
        conv_block(128),
        tf.keras.layers.Dropout(0.2),
        conv_block(256),
        tf.keras.layers.Dropout(0.2),
        
        tf.keras.layers.Flatten(),
        # bn + droput을 사용한 dense block
        dense_block(512, 0.7),
        dense_block(128, 0.5),
        dense_block(64, 0.3),
        
        tf.keras.layers.Dense(1, activation = 'sigmoid')
    ])
    return model

# 모델 2
def build_model2():
    model = tf.keras.Sequential([
        input_block(),
        
        conv_block(32),
        conv_block(64),
        conv_block(128),
        # tf.keras.layers.Dropout(0.2),
        # conv_block(256),
        # tf.keras.layers.Dropout(0.2),
        # 모델 단순화를 위해 상기 레이어를 생략
        
        tf.keras.layers.Flatten(),
        # dropout만 사용한 dense block2
        dense_block2(512, 0.7),
        dense_block2(128, 0.5),
        dense_block2(64, 0.3),
        
        tf.keras.layers.Dense(1, activation = 'sigmoid')
    ])
    return model
```

원본 데이터 모델 학습 및 평가

```python
# none augmentation

epochs = 20

with tf.device('/GPU:0'):
    model1 = build_model()

    METRICS = [
        'accuracy',
        tf.keras.metrics.Precision(name = 'precision'),
        tf.keras.metrics.Recall(name = 'recall')
    ]
    
    model1.compile(
        optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-6),
        loss = 'binary_crossentropy',
        metrics = METRICS
    )
    
    history1 = model1.fit(
        train_ds,
        steps_per_epoch = COUNT_TRAIN_TOTAL // batch_size,
        epochs = epochs,
        validation_data = val_ds,
        validation_steps = COUNT_VAL_TOTAL // batch_size,
        class_weight = class_weight,
        verbose = 0
    )
    
    model2 = build_model2()
    
    model2.compile(
        optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-6),
        loss = 'binary_crossentropy',
        metrics = METRICS
    )
    
    history2 = model2.fit(
        train_ds,
        steps_per_epoch = COUNT_TRAIN_TOTAL // batch_size,
        epochs = epochs,
        validation_data = val_ds,
        validation_steps = COUNT_VAL_TOTAL // batch_size,
        class_weight = class_weight,
        verbose = 0
    )
```

```python
# 결과 시각화
for hist in [history1, history2]:
    fig, ax = plt.subplots(1, 4, figsize=(20, 3))
    ax = ax.ravel()

    for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
        ax[i].plot(hist.history[met])
        ax[i].plot(hist.history['val_' + met])
        ax[i].set_title('Model {}'.format(met))
        ax[i].set_xlabel('epochs')
        ax[i].set_ylabel(met)
        ax[i].legend(['train', 'val'])
```

```python
test_ds = test_ds.batch(batch_size)

# 모델 평가
print("Model 1")
loss, accuracy, precision, recall = model1.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}\n')

print("Model 2")
loss, accuracy, precision, recall = model2.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}\n')
```

Augementation 데이터 모델 학습 및 평가

```python
# augmentation
with tf.device('/GPU:0'):
    # aug1 - flip up and down
    model1_1 = build_model()
    model1_1.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-6), loss = 'binary_crossentropy', metrics = METRICS)
    
    history1_1 = model1_1.fit(
        train_ds_aug1,
        steps_per_epoch = COUNT_TRAIN_TOTAL // batch_size,
        epochs = epochs,
        validation_data = val_ds,
        validation_steps = COUNT_VAL_TOTAL // batch_size,
        class_weight = class_weight,
        verbose = 0
    )
    
    # aug2 - contrast
    model1_2 = build_model()
    model1_2.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-6), loss = 'binary_crossentropy', metrics = METRICS)
    
    history1_2 = model1_2.fit(
        train_ds_aug2,
        steps_per_epoch = COUNT_TRAIN_TOTAL // batch_size,
        epochs = epochs,
        validation_data = val_ds,
        validation_steps = COUNT_VAL_TOTAL // batch_size,
        class_weight = class_weight,
        verbose = 0
    )
    
    # aug3 - both
    model1_3 = build_model()
    model1_3.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-6), loss = 'binary_crossentropy', metrics = METRICS)
    
    history1_3 = model1_3.fit(
        train_ds_aug3,
        steps_per_epoch = COUNT_TRAIN_TOTAL // batch_size,
        epochs = epochs,
        validation_data = val_ds,
        validation_steps = COUNT_VAL_TOTAL // batch_size,
        class_weight = class_weight,
        verbose = 0
    )
    
    # aug 1 - flip up and down
    model2_1 = build_model2()
    model2_1.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-6), loss = 'binary_crossentropy', metrics = METRICS)
    
    history2_1 = model2_1.fit(
        train_ds_aug1,
        steps_per_epoch = COUNT_TRAIN_TOTAL // batch_size,
        epochs = epochs,
        validation_data = val_ds,
        validation_steps = COUNT_VAL_TOTAL // batch_size,
        class_weight = class_weight,
        verbose = 0
    )
    
    # aug 2 - contrast
    model2_2 = build_model2()
    model2_2.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-6), loss = 'binary_crossentropy', metrics = METRICS)
    
    history2_2 = model2_2.fit(
        train_ds_aug2,
        steps_per_epoch = COUNT_TRAIN_TOTAL // batch_size,
        epochs = epochs,
        validation_data = val_ds,
        validation_steps = COUNT_VAL_TOTAL // batch_size,
        class_weight = class_weight,
        verbose = 0
    )
    
    # aug 3 - both
    model2_3 = build_model2()
    model2_3.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 1e-6), loss = 'binary_crossentropy', metrics = METRICS)
    
    history2_3 = model2_3.fit(
        train_ds_aug3,
        steps_per_epoch = COUNT_TRAIN_TOTAL // batch_size,
        epochs = epochs,
        validation_data = val_ds,
        validation_steps = COUNT_VAL_TOTAL // batch_size,
        class_weight = class_weight,
        verbose = 0
    )
```

```python
# 결과 시각화
for hist in [history1_1, history1_2, history1_3, history2_1, history2_2, history2_3]:
    fig, ax = plt.subplots(1, 4, figsize=(20, 3))
    ax = ax.ravel()

    for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
        ax[i].plot(hist.history[met])
        ax[i].plot(hist.history['val_' + met])
        ax[i].set_title('Model {}'.format(met))
        ax[i].set_xlabel('epochs')
        ax[i].set_ylabel(met)
        ax[i].legend(['train', 'val'])
```

```python
# 모델 평가
print("Model 1, Data Augmentation 1(flip up down)")
loss, accuracy, precision, recall = model1_1.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}\n')

print("Model 1, Data Augmentation 2(contrast)")
loss, accuracy, precision, recall = model1_2.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}\n')

print("Model 1, Data Augmentation 3(both)")
loss, accuracy, precision, recall = model1_3.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}\n')

print("Model 2, Data Augmentation 1(flip up down)")
loss, accuracy, precision, recall = model2_1.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}\n')

print("Model 2, Data Augmentation 2(contrast)")
loss, accuracy, precision, recall = model2_2.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}\n')

print("Model 2, Data Augmentation 3(both)")
loss, accuracy, precision, recall = model2_3.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}')
```

```python
print("Best Score")
loss, accuracy, precision, recall = model2_2.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}\n')
```
